#### Loc y Iloc
Pandas proporciona algunos atributos especiales indexer que exponen explicitamente ciertos esquemas de indexación.

Estos no son métodos funcionales, sino atributos que exponen una interfaz de corte particular a los datos.

Primero, el atributo `loc` permite indexar y segmentar que siempre hace referencia al índice explicito.

El atributo `iloc` permite indexar y segmentar que siempre hace referencia al indice implícito de estilo Python


````python
titanic.loc[10:15,'PassengerId':'Age']#loc puedes llamar si conoces el nombre
````

````python
titanic.iloc[10:15,0:5]#iloc con un numero
````

#### Sección de Datos

````python
titanic[10:15]#retorna las filas 10 hasta la 14
````

````python
titanic[["PassengerId","Age"]][10:15]#[[lista de columnas]][filas]
````

#### Concat
Pandas tiene una función, `pd.concat()`, que tiene una sintaxis similar a `np.concatenate` pero contiene diferente argumentos.

`pd.concat()` puede ser usado para una concatenación simple de `Series` o `DataFrames`:

````python
df1 = titanic[(titanic.Age > 18) & (titanic["Age"] < 50)][10:15][["PassengerId","Survived"]]
df1.reset_index(inplace=True)#Reseteo el index
df1
````

````python
df2 = titanic.iloc[10:15,2:5]# la que ya conocemos
df2.reset_index(inplace=True)
df2
````

````python
pd.concat([df1,df2],sort=True,axis=1)#sort = ascendente, axis=uno al lado del otro
````

````python
pd.concat([df1,df2],ignore_index=True,sort=True,axis=0)#un index continuo, ascendente, forma uno sobre el otro
````

#### Merge and join
Ambas funciones permiten que los datos de diferentes dataframes se combinen en uno solo de acuerdo con una regla de "cruce" o "busqueda".

Aunque tanto `merge` como join hacen cosas similares, la forma en que lo hacen es diferente.

La función `merge`es la función predeterminada de pandas para unir datos. Básicamente es contraparte de pandas de la union de SQL, y requiere la especificación de qué columnas de ambos dataframe se comparán. A merge no le importa en absoluto los índices definidos en ellos.

Por otro lado, la función `join` de Panda es mas conveniente (incluso utiliza merge internamente), unir es básicamente hacer una fusión aprovechando los índices de ambos marcos de datos.

La siguiente figura resume los diferentes 4 tipos de combinaciones: inner, outer, left and right.

![MERGE](http://www.datasciencemadesimple.com/wp-content/uploads/2017/09/join-or-merge-in-python-pandas-1.png)

La función merge también esta disponible como método en la clase `DataFrame`. La sintaxis básica es:

```
new_joined_df = df.merge(another_df, left_on="col_in_df", right_on="col_in_another_df",
                            how="inner"|"left"|"right|"outer")
```

El primer argumento (`another_df`), asi como `left_on` y `right_on` son argumentos obligatorios.
`left_on` especifica un nombre de columna en el dataframe `df` cuyos valores deben coincidir con los de la columna `another_df` especificados en `right_on`.

El argumento `how` es opcional y por defecto es `inner`.

````python
personal_df = pd.DataFrame([
    {"Nombre":"Alejandro", "Rol":"Director de laboratorio"},
    {"Nombre":"Jhon", "Rol":"Tecnico Laboratorio"},
    {"Nombre":"Valentina", "Rol":"Aprendiz"},
])
personal_df
````

````python
personal_df = personal_df.set_index("Nombre")
personal_df
````

````python
estudiante_df = pd.DataFrame([
    {"Nombre":"Alejandro","Carrera":"Arquitecto"},
    {"Nombre":"Jhon","Carrera":"Electronico"},
    {"Nombre":"David","Carrera":"Administración"}
])
estudiante_df=estudiante_df.set_index("Nombre")
estudiante_df
````

````python
pd.merge(personal_df,estudiante_df,left_index=True,right_index=True)#Retorna las tablas que se conectan
````

````python
pd.merge(personal_df,estudiante_df,how="outer",left_index=True,right_index=True)#una sola tabla y los vacios con NAN
````

````python
pd.merge(personal_df,estudiante_df,how="left",left_index=True,right_index=True)#una sola tabla con personal
````

````python
pd.merge(personal_df,estudiante_df,how="right",left_index=True,right_index=True)#una sola tabla con estudiante
````

````python
personal_df = personal_df.reset_index()
estudiante_df = estudiante_df.reset_index()
````

````python
pd.merge(personal_df,estudiante_df,how="left",left_on="Nombre",right_on="Nombre")
````

````python
pd.merge(personal_df,estudiante_df,how="left",on="Nombre")
````

````python
personal_df = pd.DataFrame([
    {"Primer Nombre":"Alejandro", "Apellido":"Hernandez","Rol":"Director de laboratorio"},
    {"Primer Nombre":"Jhon", "Apellido":"Cruz","Rol":"Tecnico Laboratorio"},
    {"Primer Nombre":"Valentina", "Apellido":"Ramirez","Rol":"Aprendiz"},
])
````

````python
estudiante_df = pd.DataFrame([
    {"Primer Nombre":"Alejandro","Apellido":"Hernandez","Carrera":"Arquitecto"},
    {"Primer Nombre":"Jhon","Apellido":"Cruz","Carrera":"Electronico"},
    {"Primer Nombre":"David","Apellido":"Contreras","Carrera":"Administración"}
])
````

````python
pd.merge(personal_df,estudiante_df,how="inner",left_on=["Primer Nombre","Apellido"],right_on=["Primer Nombre","Apellido"])
````
